To start, we first import needed python packages (pandas, matplotlib, numpy, etc.)

```python
import os 
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import openpyxl
```

This script requires two input arguments, EXP ID and LOPAP calibration ID.
Plots of the experiment with EXP ID will beproduced and HONO flux will be calibrated based on LOPAP cal ID. 

```python
ExpID = 68 #ID of experiments 
lopapCalID = 19 #ID of LOPAP calibration 
```

We have a measurement log file that summarise information of each experiment. 
Here we separate the excel sheets to two dataframes, experiement log, LOPAP calibration log

```python
PATH = '.' #relative path from the location of this notebook script.
#Load measuremen log files
wb_logs = openpyxl.load_workbook(os.path.join(PATH, "Experiments", "Gas_exchange", "measurement_log.xlsx"), data_only = True)
df_meas_log = pd.DataFrame(wb_logs['experiments'].values)
df_meas_log = df_meas_log.rename(columns=df_meas_log.iloc[0]).drop(0,axis=0).set_index("Exp ID")
#Load LOPAP calibration log 
df_lopap_log = pd.DataFrame(wb_logs["LOPAP calibration"].values)
df_lopap_log = df_lopap_log.rename(columns=df_lopap_log.iloc[0]).drop(0,axis=0)
```

```python
# To overview the loaded data 
df_meas_log
```

```python
# list of items
list(df_meas_log)
```

```python
# variables need for calibrationLOPAP calibration 
nitriteC = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['standard conc.'].item() #concentration of standarad solution in the unit of mg/L
dmin = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['R1 flask fill'].item().minute
dsec = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['R1 flask fill'].item().second
liquid_flow_1 = 5/(dmin+dsec/60) #5ml flow per dmin+dsec/60 min
dmin = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['R2 flask fill'].item().minute
dsec = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['R2 flask fill'].item().second    
liquid_flow_2 = 5/(dmin+dsec/60) #5ml flow per dmin+dsec/60 min
gas_flow = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['LOPAP flow'].item() #ml/min
relative_error = 0.1
```

```python
# Load time series data file that includes the perieod of LOPAP calibration  
fn = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['LOPAP data file name'].item()
# start and end of calibration 
t0 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['Start'].item()+pd.DateOffset(minutes=-20) # buffer of 20 mins from the starting point
t1 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['End'].item()+pd.DateOffset(minutes=+60)  # post-buffer of 60 mins from the end point
df_lopap = pd.read_csv(os.path.join(PATH, "Experiments", "Gas_exchange", "data", fn), sep="\t", index_col=0)
df_lopap = df_lopap.set_index(pd.to_datetime(df_lopap.index, dayfirst=True))
df_hono_cal = df_lopap.iloc[:,df_lopap.columns.get_loc("date_nm"):df_lopap.columns.get_loc("byte0_nm")] #selelct only LOPAP data (removing other concatenated data)
df_hono_cal['time'] = (df_hono_cal.index - df_hono_cal.index[0]).astype('timedelta64[m]') 
df_hono_cal = df_hono_cal.iloc[:,2:]
df_hono_cal = df_hono_cal.groupby(df_hono_cal.index).mean()
```

```python
# plot the signals from all channels during the LOPAP calibration
df_hono_cal.loc[t0:t1].iloc[:,:-1].plot(legend=False) 
plt.title('All channels - Lopap calibration ID: %d'%lopapCalID)
#plt.savefig(os.path.join(PATH, "Experiments", "Gas_exchange", "data","Figures",'Figure_signals_'+'LopapCalID_%d'%lopapCalID+'.png'),dpi=150)
```

```python
# Here, plot the time series and decide 6 time points, 
# zero air (before) [t0_1, t0_2], 
# hono_air(during calibration) [t_cal_1:t_cal_2], 
# zero_air(after)[t1_1:t1_2] 

t0_1 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['pre zero air start'].item()+pd.DateOffset(minutes=+30)
t0_2 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['pre zero air end'].item()+pd.DateOffset(minutes=+15)
t_cal_1 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['Start'].item()+pd.DateOffset(minutes=+30)
t_cal_2 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['End'].item()+pd.DateOffset(minutes=+15)
t1_1 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['post zero air start'].item()+pd.DateOffset(minutes=+35)
t1_2 = df_lopap_log[df_lopap_log["ID"]==lopapCalID]['post zero air end'].item()+pd.DateOffset(minutes=+5)

df_zero_before = df_hono_cal.loc[t0_1:t0_2]
df_zero_after = df_hono_cal.loc[t1_1:t1_2]
df_zero = pd.concat([df_zero_before,df_zero_after])
#df_zero.iloc[:,:-1].plot(marker='o',legend=False)
#plt.title('signals for zero air')

df_hono = df_hono_cal.loc[t_cal_1:t_cal_2]
#df_hono.iloc[:,:-1].plot(marker='o',legend=False)
#plt.title('signals for standard solutions')

df_hono_cal = df_hono_cal.loc[t0_1:t1_2]
df_hono_cal.iloc[:,:-1].plot(legend=False)
plt.axvspan(t0_1, t0_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t_cal_1, t_cal_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t1_1, t1_2, color='gray', alpha=0.2, lw=0)
plt.title('signals before time drift correction')
```

Often time drift correction is not nececssary for a calibration with a standard solution during a short duration 
However, here we show an example of time drift correction, which can be used for data processing.
Here, a linear function is used for the time drift correction

Also, this process includes the zero-air correction, which means that the signals are lineraly shifted to indicate zero signal for zero-air

```python
#Expand data with corrected and error per wavelength
list_corr = []

for idx1 in np.arange(0,29,1):
    try:
        fit = np.polyfit(df_zero['time'], df_zero.iloc[:,idx1], 1)
        fit_fn = np.poly1d(fit)
        df_hono_cal[df_zero.columns[idx1]+'_corr'] = df_hono_cal.iloc[:,idx1] - fit_fn(df_hono_cal['time'])
        df_hono_cal[df_zero.columns[idx1]+'_err'] = relative_error*df_hono_cal[df_zero.columns[idx1]+'_corr'] + 2*df_zero.iloc[:,idx1].std() #relative error + noise
    except:
        print(list(df_zero)[idx1])

list_corr = df_hono_cal.columns[df_hono_cal.columns.str.contains("_corr")]
list_err = df_hono_cal.columns[df_hono_cal.columns.str.contains("_err")]

df_hono_cal.loc[:,list_corr].plot(legend=False)
plt.axvspan(t0_1, t0_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t_cal_1, t_cal_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t1_1, t1_2, color='gray', alpha=0.2, lw=0)
plt.title('zero air corrected signals')

# plotting with errorbar
#for idx1 in np.arange(0,29,1):
#    plt.errorbar(df_hono_cal.index,df_hono_cal.loc[:,list_corr[idx1]],yerr=df_hono_cal.loc[:,list_err[idx1]])
```

```python
# for a calibration with the standard solution: representative values for zero air (no HONO) and hono air (air with known concentraion of HONO - driven from the standard solution used)      
ch1_ref = nitriteC/1000*liquid_flow_1/1000/46*6.022E+23/(24600000000000000000*gas_flow)*1000000000000 #HONO concentration [ppbv] detected in ch1 
ch2_ref = nitriteC/1000*liquid_flow_2/1000/46*6.022E+23/(24600000000000000000*gas_flow)*1000000000000 #HONO concentration [ppbv] detected in ch2

zero_air = df_hono_cal.loc[df_zero.index].median() #Here, I muse median values from the time series. using a mean can be an alternative
hono_air = df_hono_cal.loc[df_hono.index].median()
```

```python
# signal to real HONO values 

for idx1 in np.arange(1,15,1):
    chn = df_zero.columns[idx1]+'_corr'
    fit = np.polyfit([zero_air[chn],hono_air[chn]], [0,ch1_ref], 1)
    fit_fn1 = np.poly1d(fit)
    plt.plot([zero_air[chn],hono_air[chn]], [0,ch1_ref],'o',label=df_zero.columns[idx1])
    
for idx1 in np.arange(16,29,1):
    chn = df_zero.columns[idx1]+'_corr'
    fit = np.polyfit([zero_air[chn],hono_air[chn]], [0,ch2_ref], 1)
    fit_fn2 = np.poly1d(fit)
    plt.plot([zero_air[chn],hono_air[chn]], [0,ch2_ref],'o',label=df_zero.columns[idx1])
    
plt.legend()    
plt.xlabel('ABS')    
plt.ylabel('HONO [pptV]')   
plt.show()
```

```python
# Signals calibrated to the real HONO values 
s_eff = 0.998 # sampling efficiency correction: this can be a fion of gas and liquid flows, but here we assumed a constant

df_hono = []
for idx1, idx2 in zip([4,5,6,7,8,10,11],[19,20,21,22,23,24,25]): #Here we already selected a few wavelengths that can be sued for HONO detection
    chn = df_zero.columns[idx1]+'_corr'
    fit = np.polyfit([zero_air[chn],hono_air[chn]], [0,ch1_ref], 1)
    fit_fn1 = np.poly1d(fit)
    chn2 = df_zero.columns[idx2]+'_corr'
    fit = np.polyfit([zero_air[chn2],hono_air[chn2]], [0,ch2_ref], 1)
    fit_fn2 = np.poly1d(fit)
    temp = 0.001*(fit_fn1(df_hono_cal[chn])) 
    temp2 = 0.001*(fit_fn2(df_hono_cal[chn2]))
    dftemp0 = pd.DataFrame(temp/s_eff-(temp2+temp*(1-s_eff)), index=df_hono_cal.index,columns={'HONO_'+df_zero.columns[idx1]})
    df_hono.append(dftemp0)
    
df_hono = pd.concat(df_hono,axis=1)
df_hono.plot()
plt.axvspan(t0_1, t0_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t_cal_1, t_cal_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t1_1, t1_2, color='gray', alpha=0.2, lw=0)
plt.xlabel('time')    
plt.ylabel('HONO [ppbV]')   
plt.grid()
```

## Till now, we have prepared a calibration of LOPAP signals for HONO emission measurements
## From here, we will load all time series data and plot 

```python
# Load vbus data (which is a tabular data prepared in the experimental setup, providing concatenate data of LICOR, Picarro, LOPAP, and other sensors.
if df_meas_log.VBUS.iloc[ExpID]:
    fnames = glob.glob(os.path.join(PATH, "Experiments", "Gas_exchange", "data", df_meas_log['VUBS data file name'].iloc[ExpID]))
    df_vbus = pd.read_csv(fnames[0], sep="\t", index_col=0)
    df_vbus = df_vbus.set_index(pd.to_datetime(df_vbus.index, dayfirst=True)) #set inddex with the measurement time wiht datetime type, which is convenient for aggregating data in time
    df_vbus = df_vbus.loc[:, (df_vbus != 0).any(axis=0)] #remove columns with only zero values (such as entries from Picarro. which is not connected to the setup)
else:
    print('no vbus file available')
```

```python
# Load data from the NOx analyzer
if df_meas_log.Teledyne.iloc[ExpID]:
    fnames = glob.glob(os.path.join(PATH, "Experiments", "Gas_exchange", "data", df_meas_log['T200UP data file name'].iloc[ExpID]))
    fn_nox =[]
    for fn in fnames:
        df_part = pd.read_csv(fn, sep="\t", index_col=0)       
        fn_nox.append(df_part.set_index(pd.to_datetime(df_part.index, format='%d.%m.%Y %H:%M:%S')+pd.DateOffset(minutes=0)))      
    df_nox = pd.concat(fn_nox)
    df_nox = df_nox.loc[:, (df_nox != 0).any(axis=0)]
else:
    print('no teledyne file available')      
```

```python
df_nox
```

```python
df_hono = df_vbus.time_nm
for idx1, idx2 in zip([4,5,6,7,8,10,11],[19,20,21,22,23,24,25]):
    chn = df_zero.columns[idx1]+'_corr'
    fit = np.polyfit([zero_air[chn],hono_air[chn]], [0,ch1_ref], 1)
    fit_fn1 = np.poly1d(fit)
    chn2 = df_zero.columns[idx2]+'_corr'
    fit = np.polyfit([zero_air[chn2],hono_air[chn2]], [0,ch2_ref], 1)
    fit_fn2 = np.poly1d(fit)
    temp = 0.001*(fit_fn1(df_vbus[df_zero.columns[idx1]])) 
    temp2 = 0.001*(fit_fn2(df_vbus[df_zero.columns[idx2]]))
    dftemp0 = pd.DataFrame(temp-temp2, index=df_vbus.index,columns={'HONO_'+df_zero.columns[idx1]})
    df_hono = pd.concat([df_hono,dftemp0],axis=1)

    
df_hono['corr_time'] = df_hono.index + pd.DateOffset(minutes=-12)
df_hono = df_hono.reset_index().set_index(df_hono['corr_time'])
df_hono['time'] = (df_hono.index - df_hono.index[0]).astype('timedelta64[m]') 

# time drift correction    

#t0_1 = df_meas_log['pre zero air start'].iloc[ExpID]+pd.DateOffset(minutes=0) # These are considered 
#t0_2 = df_meas_log['pre zero air end'].iloc[ExpID]+pd.DateOffset(minutes=0)
t0_1 = df_meas_log['post zero air start'].iloc[ExpID-1]+pd.DateOffset(minutes=+5)
t0_2 = df_meas_log['post zero air end'].iloc[ExpID-1]+pd.DateOffset(minutes=-5)

t1_1 = df_meas_log['post zero air start'].iloc[ExpID]+pd.DateOffset(minutes=+5)
t1_2 = df_meas_log['post zero air end'].iloc[ExpID]+pd.DateOffset(minutes=-5)

df_zero_corr_before = df_hono.loc[t0_1:t0_2]
df_zero_corr_after = df_hono.loc[t1_1:t1_2]
df_zero_corr = pd.concat([df_zero_corr_before,df_zero_corr_after])

list_cols = df_hono.columns[df_hono.columns.str.contains("HONO_")]

for coln in list(list_cols):
    fit = np.polyfit(df_zero_corr['time'], df_zero_corr.loc[:,coln], 1)
    fit_fn = np.poly1d(fit)
    df_hono[coln+'_corr'] = df_hono.loc[:,coln] - fit_fn(df_hono['time'])
    df_hono[coln+'_err'] = 0.1*df_hono[coln+'_corr'] + 2*df_zero_corr.loc[:,coln].std() #relative error + noise

list_cols = df_hono.columns[df_hono.columns.str.contains("_corr")]
df_hono.loc[t0_1:t1_2,list_cols].plot()
plt.axvspan(t0_1, t0_2, color='gray', alpha=0.2, lw=0)
plt.axvspan(t1_1, t1_2, color='gray', alpha=0.2, lw=0)
plt.xlabel('time')    
plt.ylabel('HONO [ppbV]')   
plt.grid()
```

```python
#  reindex and concat data, to match the time index, we resample by 60s
df_vbus_m = df_vbus.resample('60s').mean()
df_nox_m = df_nox.resample('60s').mean()
t0 = df_meas_log['Start'].iloc[ExpID]+pd.DateOffset(minutes=0)
t1 = df_meas_log['End'].iloc[ExpID]+pd.DateOffset(minutes=0)
df_vbus_part = df_vbus_m.loc[t0:t1]
df_nox_reindex = df_nox_m.reindex(df_vbus_part.index, method="nearest") # reindex the nox data with the same index with df_vbus (second differences are ignored)
df_hono_reindex = df_hono.reindex(df_vbus_part.index, method="nearest")
df = pd.concat([df_vbus_part,df_hono_reindex,df_nox_reindex],axis=1)
```

```python
elaps_time = (df.index - df.iloc[df.index.get_loc(t0,method='nearest')].name).astype('timedelta64[m]')
df['elaps_time'] = elaps_time/60 #add a column of elapsed time from the start of the experiment
try:
    h2o_cum = (df["H2OBmmol"]-df["H2OAmmol"]).cumsum()
except KeyError:
    h2o_cum = (df["H2OBmmol"]).cumsum() # when df["H2OAmmol"] is omitted at the beginning due to all-zero entries 

lost_water = h2o_cum*df_meas_log.m_dif.iloc[ExpID]/h2o_cum[-1]
if ExpID == 12:
    df['lost_water'] = 3.3*0.997-lost_water
else:
    try:
        df['lost_water'] = df_meas_log.added_volume.iloc[ExpID]*0.997-lost_water #volume times water density 
    except TypeError:
        df['lost_water'] = df_meas_log.m_dif.iloc[ExpID]-lost_water

#df.to_csv(os.path.join(PATH, "Experiments", "Gas_exchange", "data","Processed_data",'Data_'+filename+'.csv'))
```

```python
xelaps = True # plot data with elapsed hours on the x-axis, if false, plots with the real time 
if xelaps:
    df2 = df.reset_index(drop=True).set_index(df['elaps_time'])
else:
    df2 = df

fig, axs = plt.subplots(3, 1, sharex=True, figsize=(4.5,6), dpi=150)
fig.subplots_adjust(hspace=0.1)# Remove horizontal space between axes

try:
    axs[0].plot(df2["H2OBmmol"]-df2["H2OAmmol"], label="$\Delta$ H$_2$O", color ="k")
except KeyError:
    axs[0].plot(df2["H2OBmmol"], label="H$_2$O", color ="tab:blue")
axs[0].set_ylim(-0.1,8)
axs[0].set_ylabel("$\Delta$ H$_2$O [mmol]",fontsize=12)

try:
    ax3=axs[0].twinx()
    ax3.plot(df2['Hum1'],ls='--', label="inflow", color ="tab:blue")
    ax3.plot(df2['Hum2'],ls=':', label="outflow", color ="tab:blue")
    ax3.set_ylabel("Humidiy [%]",fontsize=12,color="tab:blue")
    ax3.set_ylim(-0.5, 100)
    ax3.legend(fontsize=8,loc="upper right")
except:
    pass

axs[1].plot(df2["NO"],color ="tab:green")
axs[1].set_yticks(np.arange(0, 2, 0.4))
axs[1].set_ylim(-0.1,2)
axs[1].set_ylabel("NO [ppb]",fontsize=12,color ="tab:green")

ax32=axs[1].twinx()
ax32.plot(df2["NO2"],color ="tab:orange")
ax32.set_ylabel("NO$_2$ [ppb]",fontsize=12,color ="tab:orange")
ax32.set_ylim(-0.1, 2)
ax32.set_yticks(np.arange(0,2, 0.4))

if not df_meas_log.LOPAP.iloc[ExpID]:
    print('no calibrated HONO available')
else:

    axs[2].plot(df2["HONO_538.1_nm_corr"],label='HONO (540nm)',color ="tab:blue")
    axs[2].plot(df2["HONO_548.0_nm_corr"],label='HONO (550nm)',color ="tab:purple")
    axs[2].plot(df2["HONO_567.7_nm_corr"],label='HONO (568nm)',color ="tab:green")
    axs[2].plot(df2["HONO_577.5_nm_corr"],label='HONO (580nm)',color ="tab:red")
    axs[2].plot(df2["HONO_587.7_nm_corr"],label='HONO (588nm)',color ="tab:orange")
    axs[2].legend(fontsize=8, loc='upper left')
    axs[2].set_yticks(np.arange(0, 0.5, 0.1))
    axs[2].set_ylim(-0.01, 0.5)
    axs[2].set_ylabel("HONO [ppbV]",fontsize=12)


ax2=axs[2].twinx()
ax2.plot(df2['lost_water']+0.5,ls='--',color ="blue") # Here the correction factore 0.5 is given, for plotting. The amount of lost water can be recalculated
ax2.set_ylabel("H$_2$O [g]",fontsize=12,color="blue")
ax2.set_ylim(-0.2,2.2)
ax2.set_yticks(np.arange(0,3, 1))

if not xelaps:
    axs[2].set_xlabel("Time",fontsize=12)
    axs[2].tick_params(axis="x",rotation=45)
else:
    axs[2].set_xlabel("Time [hour]",fontsize=12)
    #plt.xlim(0,df2.elaps_time.max())
    plt.xlim(0,5)
    
try:
    titlename = 'ExpID: %d, %d-day-old AOB culture \n OD$_{600}$: %1.4f, pH: %1.1f, NO$_2^-$: %dµM)'%(ExpID, df_meas_log.loc[ExpID]['day'], df_meas_log.loc[ExpID]['OD600'],df_meas_log.loc[ExpID]['pH'],df_meas_log.loc[ExpID]['NO2- (µM)'])
except:
    titlename = 'ExpID: %d '%ExpID + df_meas_log.loc[ExpID]['remark']

plt.suptitle(titlename, fontsize=12)
plt.tight_layout()
plt.show()    
#plt.savefig(os.path.join(PATH, "Experiments", "Gas_exchange", "data","Figures",'Figure_'+filename+'.png'),format="png")
```
